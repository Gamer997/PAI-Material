### Kind
* 'line' (default): Line plot.
* 'bar': Vertical bar plot.
* 'barh': Horizontal bar plot.
* 'hist': Histogram.
* 'box': Box plot.
* 'kde': Kernel Density Estimation plot.
* 'density': Same as 'kde'.
* 'area': Area plot.
* 'scatter': Scatter plot.
* 'hexbin': Hexagonal bin plot.

### Marker
* 'o': Circle marker (default).
* 's': Square marker.
* '^': Triangle marker.
* '*': Star marker.
* '+': Plus marker.
* 'x': X marker.
* 'D': Diamond marker.
* '.': Point marker.
