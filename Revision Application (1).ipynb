```python
import tkinter as tk
from tkinter import filedialog, ttk
import cv2
from PIL import Image, ImageTk
import numpy as np
from pydub import AudioSegment
import simpleaudio as sa
```

## Image processing part
Look at edge detection function

```python
class ImageProcessingApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Image Processing App")

        # Variables
        self.image_path = None
        self.original_image = None

        # Widgets
        self.upload_button = tk.Button(self.master, text="Upload Image", command=self.upload_image)
        self.upload_button.pack(pady=10)

        self.gaussian_button = tk.Button(self.master, text="Gaussian Blur", command=self.apply_gaussian_blur)
        self.gaussian_button.pack(pady=5)

        self.edge_button = tk.Button(self.master, text="Edge Detection (Sobel)", command=self.apply_edge_detection)
        self.edge_button.pack(pady=5)

        self.canny_button = tk.Button(self.master, text="Canny Edge Detector", command=self.apply_canny)
        self.canny_button.pack(pady=5)

        self.image_label = tk.Label(self.master)
        self.image_label.pack(pady=10)

        self.back_button = tk.Button(self.master, text="Back to Main Window", command=self.show_main_window)
        self.back_button.pack(pady=10)

    def upload_image(self):
        file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.png;*.jpg;*.jpeg")])
        if file_path:
            self.image_path = file_path
            self.original_image = cv2.imread(self.image_path)
            self.update_image_label(self.original_image)

    def update_image_label(self, image):
        # Resize the image for better visualization
        resized_image = cv2.resize(image, (300, 200))
        resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(resized_image)
        img = ImageTk.PhotoImage(img)
        self.image_label.config(image=img)
        self.image_label.image = img

    def apply_gaussian_blur(self):
        if self.original_image is not None:
            # Convert the image to grayscale if necessary
            gray_image = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2GRAY) if len(self.original_image.shape) == 3 else self.original_image

            # Create a 5x5 Gaussian kernel
            gaussian_kernel = cv2.getGaussianKernel(5, 5)

            # Apply Gaussian blur using filter2D
            blurred_image = cv2.filter2D(gray_image, -1, gaussian_kernel)

            # Update the image label with the blurred image
            self.update_image_label(blurred_image)

    def apply_edge_detection(self):
        if self.original_image is not None:
            gray_image = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2GRAY)

            # Use Sobel operator for edge detection
            sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
            sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)

            edge_image = np.sqrt(sobel_x**2 + sobel_y**2)
            edge_image = np.uint8(edge_image)

            self.update_image_label(edge_image)

    def apply_canny(self):
        if self.original_image is not None:
            # Convert the image to grayscale
            gray_image = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2GRAY)

            # Apply Canny edge detection
            edges = cv2.Canny(gray_image, 100, 200)  # You can adjust the threshold values (100 and 200) as needed

            # Update the image label with the Canny edge-detected image
            self.update_image_label(edges)


    def show_main_window(self):
        # Close the current window
        self.master.destroy()

        # Create and show the main window
        main_window = tk.Tk()
        main_app = MainWindow(main_window)
        main_window.mainloop()

```
