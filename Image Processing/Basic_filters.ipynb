```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r"SunFlower_Image.jpg", 1)
# Loading the image

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))

stretch_near = cv2.resize(image, (780, 540), 
			interpolation = cv2.INTER_LINEAR)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
	plt.subplot(2, 2, i + 1)
	plt.title(Titles[i])
	plt.imshow(images[i])

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./kid.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
b, g, r = cv2.split(img)

fig, axs = plt.subplots(3, 5, figsize=(15,15))
axs = axs.flatten()
axs[0].imshow(img)
axs[0].axis('on')
axs[0].set_title('Original')

axs[1].imshow(gray, cmap='gray')
axs[1].axis('on')
axs[1].set_title('Gray Scale')

axs[2].imshow(b, cmap='Blues')
axs[2].axis('off')
axs[2].set_title('Blue')

axs[3].imshow(g, cmap='Greens')
axs[3].axis('off')
axs[3].set_title('Green')

axs[4].imshow(r, cmap='Reds')
axs[4].axis('off')
axs[4].set_title('Red')

axs[5].imshow(cv2.GaussianBlur(img, (11, 11), 0))
axs[5].axis('off')
axs[5].set_title('Gaussian Blur')

axs[6].imshow(cv2.medianBlur(img, 7))
axs[6].axis('off')
axs[6].set_title('Median Blur')

axs[7].imshow(cv2.bilateralFilter(img, 9, 75,75))
axs[7].axis('off')
axs[7].set_title('Bilateral Filter')

axs[8].imshow(cv2.rotate(img, cv2.ROTATE_180))
axs[8].axis('off')

axs[9].imshow(cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE))
axs[9].axis('off')

axs[10].imshow(cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE))
axs[10].axis('off')

axs[11].imshow(cv2.resize(img, (200,200)))
axs[11].axis('off')
axs[11].set_title('Resized')

axs[12].imshow(cv2.flip(img, 1))
axs[12].axis('off')
axs[12].set_title('Horizontal Flip')

axs[13].imshow(cv2.flip(img, 0))
axs[13].axis('off')
axs[13].set_title('Vertical Flip')

plt.show()
```
