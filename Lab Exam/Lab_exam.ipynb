```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread("aqsa.jpeg", 0)

# Flatten the image
flattened_image = image.flatten()

# Convert the flattened image to audio
audio_data = np.array(flattened_image, dtype=np.float32)

audio_data /= np.max(audio_data)

# Display the audio
plt.figure()
plt.plot(audio_data)
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Audio waveform of image.png")
plt.show()

# Visualize the STFT of the audio
plt.figure(figsize=(13,14))
audio_stft = librosa.stft(audio_data)
in_decibels = np.abs(librosa.amplitude_to_db(audio_stft))
librosa.display.specshow( in_decibels,sr=10000)

plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.colorbar()
plt.title("STFT of audio.png")
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = librosa.load("aqsa.jpeg", sr=10000, mono=True)

# Flatten the image
flattened_image = image.flatten()

# Convert the flattened image to audio
audio_data = np.array(flattened_image, dtype=np.float32)
audio_data /= np.max(audio_data)

# Display the audio waveform
plt.figure()
plt.plot(audio_data)
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Audio waveform of image.png")
plt.show()

# Visualize the STFT of the audio
plt.figure()
librosa.display.specshow(librosa.core.spectrum.stft(audio_data, n_fft=2048, hop_length=512), sr=10000, x_axis='time', y_axis='log')
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.colorbar()
plt.title("STFT of image.png")
plt.show()
```

```python
import cv2
import numpy as np
from scipy.io import wavfile

# Load the image
image = cv2.imread("aqsa.jpeg", cv2.IMREAD_GRAYSCALE)

# Flatten the image
flattened_image = image

# Convert the flattened image to audio
sample_rate = 10000
audio_data = np.array(flattened_image, dtype=np.float32)
audio_data /= np.max(audio_data)
wavfile.write("image.wav", sample_rate, audio_data)

# Visualize the STFT of the audio
import matplotlib.pyplot as plt

plt.figure()
plt.specgram(audio_data, Fs=sample_rate)
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.colorbar()
plt.title("STFT of image.wav")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('match.csv')

# Drop any rows with missing values
df = df.dropna()

# Create a box plot of the "Won By Deal" column
plt.figure()
plt.boxplot(df['Won_By'])
plt.xlabel('Won_By')
plt.ylabel('Value')
plt.title('Box Plot of Won By Deal')
plt.show()

# Identify outliers using IQR method
Q1 = df['Won_By'].quantile(0.25)
Q3 = df['Won_By'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers using boolean indexing
outliers = df[(df['Won_By'] < lower_bound) | (df['Won_By'] > upper_bound)]

# Print the outliers
print('Outliers:')
print(outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords corpus if not already downloaded
# import nltk
# nltk.download('stopwords')

# Read the CSV file
df = pd.read_csv('Match.csv')

# Assume your CSV has columns: 'Review', 'Positive', 'Negative', 'Stopwords'
# Modify these column names accordingly based on your CSV structure

# Function to classify sentiment based on word frequency
def classify_sentiment(review, positive_words, negative_words):
    # Tokenize and remove stopwords
    review_words = [word.lower() for word in word_tokenize(review) if word.lower() not in stop_words]
    
    # Count word occurrences
    word_counts = Counter(review_words)
    
    # Count positive and negative words
    positive_count = sum(word_counts[word] for word in positive_words if word in word_counts)
    negative_count = sum(word_counts[word] for word in negative_words if word in word_counts)
    
    # Classify sentiment
    if positive_count > negative_count:
        return 'Positive'
    elif positive_count < negative_count:
        return 'Negative'
    else:
        return 'Neutral'

# Remove duplicates and convert positive, negative, and stop words to lowercase
positive_words = set(map(str.lower, df['Positive'].dropna().unique()))
negative_words = set(map(str.lower, df['Negative'].dropna().unique()))
stop_words = set(stopwords.words('english'))
# stop_words = set(map(str.lower, df['Stopwords'].dropna().str.split(',').explode().unique()))

# Apply sentiment analysis to each review
df['Sentiment'] = df['Review'].apply(lambda x: classify_sentiment(x, positive_words, negative_words))

# Display a chart showing the distribution of sentiments
sentiment_distribution = df['Sentiment'].value_counts()
sentiment_distribution.plot(kind='bar', color=['green', 'red', 'gray'])
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.title('Sentiment Analysis of Reviews')
plt.show()

# Display the classified reviews
print(df[['Review', 'Sentiment']])
```
