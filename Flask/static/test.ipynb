```python
import numpy as np

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

# Stack arrays vertically
vertically_stacked = np.vstack((array1, array2))
print(vertically_stacked)
# Output:
# [[1 2 3]
#  [4 5 6]]
vertically_stacked.shape
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data, index=['A', 'B', 'C'])

# Using loc to access rows and columns by labels
result_loc = df.loc['B', 'Age']
print(result_loc)
# Output: 30

# Slicing using loc
sliced_df_loc = df.loc['A':'B', ['Name', 'City']]
print(sliced_df_loc)
df
# Output:
#     Name           City
# A  Alice       New York
# B    Bob  San Francisco
```

```python
import pandas as pd

# Create a DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25, 30, 35],
    'Status': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Pass'],
    'Numbers':[22, 11, 32, 13, 55,66]
}
df = pd.DataFrame(data)

# Group by 'Category'
grouped = df.groupby('Category')

# Calculate the mean value for each group
mean_values = grouped.mean()

# Display the original DataFrame and the result
print("Original DataFrame:")
print(df)

print("\nGrouped by 'Category' with Mean Values:")
print(mean_values)
df.groupby("Category")[["Value","Numbers"]].mean()
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Column1': np.random.rand(100),
    'Column2': np.random.randn(100),
    'Column3': np.random.randint(1, 100, size=100)
}
df = pd.DataFrame(data)

# Compute maximum values for each column
max_values = df.max()

# Compute standard deviations for each column
std_values = df.std()

# Compute quantiles (25%, 50%, 75%) for each column
quantiles = df.quantile([0.25, 0.5, 0.75])

# Display the results
print("Maximum Values:")
print(max_values)

print("\nStandard Deviations:")
print(std_values)

print("\nQuantiles (25%, 50%, 75%):")
print(quantiles)
```

```python
from nltk import ngrams
from nltk.tokenize import word_tokenize

# Example sentence
sentence = "This is an example sentence for generating n-grams."

# Tokenize the sentence into words
words = word_tokenize(sentence)

# Function to generate n-grams
def generate_ngrams(text, n):
    n_grams = ngrams(text, n)
    return [' '.join(gram) for gram in n_grams]

# Generate bigrams (n=2)
bigrams = generate_ngrams(words, 2)
print("Bigrams:", bigrams)

# Generate trigrams (n=3)
trigrams = generate_ngrams(words, 3)
print("Trigrams:", trigrams)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Sample text
text = "This is a sample sentence. Tokenization and removing stopwords are important preprocessing steps in NLP."

# Tokenize the text into sentences
sentences = sent_tokenize(text)

# Define and get English stopwords
stop_words = set(stopwords.words('english'))

# Tokenize each sentence into words and remove stopwords
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]
filtered_tokenized_sentences = [
    [word for word in tokens if word.lower() not in stop_words]
    for tokens in tokenized_sentences
]

# Display results
print("Original Text:")
print(text)

print("\nTokenization at Sentence Level:")
print(sentences)

print("\nTokenization at Word Level:")
print(tokenized_sentences)

print("\nTokenization with Stopword Removal:")
print(filtered_tokenized_sentences)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
        'Age': [25, 30, 22, 28, 35],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Miami', 'San Francisco']}

df = pd.DataFrame(data)

# Plotting options and explanations
# 1. Line plot of Age vs Name
df.plot(kind='line', x='Name', y='Age', title='Age by Name')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()

# 2. Histogram of Age distribution
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

# 3. Bar chart of Age grouped by City
df.groupby('City')['Age'].mean().plot(kind='bar', color=['red', 'green', 'blue', 'purple', 'orange'])
plt.xlabel('City')
plt.ylabel('Average Age')
plt.title('Average Age by City')
plt.show()

# Note: You can modify these plots and explore different options based on your specific needs.
```
